## 2.4 Встроенный тип отображения: dict

- **Словарь** (`dict`) — неупорядоченная коллекция пар `ключ: значение`.

- **Ключ**: только неизменяемый тип (int, float, str, tuple). Значение: любой объект.

- **Литерал**: {key1: val1, key2: val2}.

- **Конструктор**: `dict(one=1, two=2)` или `dict(zip([keys], [values]))`.

```python
dict1 = {'one': 1, 'two': 2}
print({1: 1, 'second': [2], (3,): {3:3}})
```

**Операции над словарями:**

1. **Индексация по ключу** [key] (для получения и изменения):

```python
dict1['one'] = 11      # изменение
print(dict1['one'])    # 11
dict1['new'] = 2       # создание нового
# dict1['two'] # KeyError!
print(dict1.get('two')) # None (безопасный доступ)
```

2. **Длина** len():

```python
print(len(dict1))
```

3. **Принадлежность** in / not in (проверка ключа):

```python
print('one' in dict1)  # True
print(11 in dict1)     # False (11 - значение)
```

4. **Объединение** | и |=:

```python
X, Y = {'one': 1}, {'one': 'updated', 'two': 2}
print(X | Y)
```

- **Кодовый трюк:** Разреженные многомерные массивы через кортежи-ключи.

```python
matrix = {}
matrix[1,10,100] = 5
print(matrix[(1,10,100)])
```

# 2.12 Файловый объект

```python
# ЗАПИСЬ В ФАЙЛ
file = open("test.txt", "w", encoding="utf-8")

# write - запись одной строки
file.write("Строка 1\n")

# writelines - запись нескольких строк
file.writelines(["Строка 2\n", "Строка 3\n"])

file.close()

# ЧТЕНИЕ ФАЙЛА
file = open("test.txt", "r", encoding="utf-8")

# read - чтение всего файла
print("read():")
print(file.read())

file.seek(0)  # возвращаем указатель в начало

# readline - чтение одной строки
print("\nreadline():")
print(file.readline(), end='')
print(file.readline(), end='')

file.seek(0)

# readlines - чтение всех строк в список
print("\n\nreadlines():")
lines = file.readlines()
print(lines)

file.close()
```

Имена переменных и функций должны писаться строчными буквами. Для разделения слов в имени используется символ подчеркивания _. Такой стиль повышает читаемость кода и является общепринятым стандартом в сообществе Python. Вопрос: Как выполняется преобразование типов при смешанных числовых операциях?

Ответ: При выполнении операций с операндами разных числовых типов Python автоматически преобразует операнды к наиболее "широкому" типу, который может представлять большее математическое множество. Порядок расширения: int - float - complex.

```python
# int + float = float
result1 = 5 + 2.5
print(f"5 + 2.5 = {result1}, тип: {type(result1)}")  # float

# int + complex = complex
result2 = 3 + (2 + 1j)
print(f"3 + (2+1j) = {result2}, тип: {type(result2)}")  # complex

# float + complex = complex
result3 = 1.5 + (2 + 3j)
print(f"1.5 + (2+3j) = {result3}, тип: {type(result3)}")  # complex

# Смешанные типы в одном выражении
result4 = 10 + 2.5 + (1 + 2j)
print(f"10 + 2.5 + (1+2j) = {result4}, тип: {type(result4)}")  # complex

# bool наследуется от int
print(f"True + 10 = {True + 10}")  # 11
```

Вопрос: Что такое ссылка на объект в Python?

Ответ: Ссылка на объект — это имя переменной, которое связывает имя с объектом в памяти. В Python переменные не содержат значения напрямую, а являются ссылками на объекты. Один объект может иметь несколько ссылок.

```python
# Создание объекта и ссылки
a = [1, 2, 3]  # a ссылается на список [1, 2, 3]
b = a          # b ссылается на тот же объект
c = [1, 2, 3]  # c ссылается на новый объект

print(f"a is b: {a is b}")  # True - одна и та же ссылка
print(f"a is c: {a is c}")  # False - разные объекты

# Изменение через одну ссылку влияет на все ссылки на этот объект
a.append(4)
print(f"a после изменения: {a}")
print(f"b после изменения a: {b}")  # b также изменился

# Проверка количества ссылок
import sys
x = [1, 2, 3]
print(f"Количество ссылок на объект: {sys.getrefcount(x) - 1}")  # -1 для учета аргумента getrefcount
```

Вопрос: В чем разница между методами find и index для строк?

Ответ: Оба метода ищут подстроку в строке и возвращают индекс первого вхождения. Разница в поведении при отсутствии подстроки: find() возвращает -1, а index() вызывает исключение ValueError.

```python
text = "Python programming is fun"

# Поиск существующей подстроки - оба метода работают одинаково
print(f"find('programming'): {text.find('programming')}")   # 7
print(f"index('programming'): {text.index('programming')}") # 7

# Поиск несуществующей подстроки
print(f"find('Java'): {text.find('Java')}")  # -1

try:
    print(f"index('Java'): {text.index('Java')}")
except ValueError as e:
    print(f"index('Java'): Ошибка - {e}")

# Использование с параметрами
print(f"find('p', 5): {text.find('p', 5)}")     # 14
print(f"find('p', 10, 15): {text.find('p', 10, 15)}")  # -1

# Практический пример с условием
def find_substring(text, substring):
    if text.find(substring) != -1:
        return f"Найдено в позиции {text.find(substring)}"
    return "Не найдено"

print(find_substring(text, "Python"))
print(find_substring(text, "Ruby"))
```

Вопрос: В чем разница между методами sort и sorted для списков?

Ответ: Метод sort() сортирует список на месте (изменяет исходный список) и возвращает None. Функция sorted() возвращает новый отсортированный список, не изменяя исходный.

```python
# Метод sort() - изменяет исходный список
numbers = [3, 1, 4, 1, 5, 9, 2]
print(f"Исходный список: {numbers}")
result = numbers.sort()
print(f"После sort(): {numbers}")
print(f"sort() возвращает: {result}")

print("-" * 40)

# Функция sorted() - создает новый список
numbers = [3, 1, 4, 1, 5, 9, 2]
print(f"Исходный список: {numbers}")
sorted_numbers = sorted(numbers)
print(f"После sorted(): {sorted_numbers}")
print(f"Исходный список не изменился: {numbers}")

print("-" * 40)

# Дополнительные параметры
words = ["banana", "apple", "cherry", "date"]

# Сортировка по длине строки
words.sort(key=len)
print(f"Сортировка по длине (sort): {words}")

# Сортировка в обратном порядке
reverse_sorted = sorted(words, reverse=True)
print(f"Обратная сортировка (sorted): {reverse_sorted}")
```

Вопрос: В чем отличие кортежа от списка?

Ответ: Основные отличия:

Изменяемость: списки изменяемы (mutable), кортежи неизменяемы (immutable)

Синтаксис: списки используют [], кортежи используют ()

Производительность: кортежи обрабатываются быстрее и занимают меньше памяти

Использование: кортежи могут быть ключами словарей, списки - нет

```python
# Изменяемость
my_list = [1, 2, 3]
my_tuple = (1, 2, 3)

my_list[0] = 10      # работает
print(f"Список после изменения: {my_list}")

try:
    my_tuple[0] = 10  # вызывает ошибку
except TypeError as e:
    print(f"Кортеж изменить нельзя: {e}")

print("-" * 40)

# Использование в качестве ключей словаря
coordinates = {(10, 20): "Point A"}  # кортеж как ключ
print(f"Кортеж как ключ: {coordinates[(10, 20)]}")

try:
    invalid_dict = {[10, 20]: "Point A"}  # список как ключ - ошибка
except TypeError as e:
    print(f"Список не может быть ключом: {e}")

print("-" * 40)

# Распаковка кортежа
point = (15, 25)
x, y = point
print(f"Распаковка кортежа: x={x}, y={y}")

# Возврат нескольких значений из функции
def get_min_max(numbers):
    return min(numbers), max(numbers)

minimum, maximum = get_min_max([4, 2, 8, 1, 9])
print(f"Минимум: {minimum}, Максимум: {maximum}")
```
